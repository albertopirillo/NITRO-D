```python
from lib.optim.lr_schedulers import ConstantLR, LinearLR, ExponentialLR, HalvingLR
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

## Auxiliary functions

```python
def plot(x, y, figsize=(16, 5)):
    fig, axs = plt.subplots(1, 1, sharey=False, figsize=figsize)
    axs.plot(x, y, linewidth=2, zorder=2)
    axs.grid(True)
    axs.set_xticks(np.arange(0, len(x), 5))
    axs.set_title('LR Schedule')
    axs.set_xlabel('Epoch')
    axs.set_ylabel('Learning rate')
```

```python
def get_schedule(scheduler, epochs):
    x = np.arange(0, epochs, 1)
    y = np.zeros(epochs)
    for i in range(epochs):
        scheduler.step()
        y[i] = scheduler.get_lr()
    return x, y
```

## ConstantLR

```python
scheduler = ConstantLR(base_lr=256)
x, y = get_schedule(scheduler, epochs=50)
```

```python
plot(x, y)
```

## LinearLR

```python
scheduler = LinearLR(base_lr=512, from_epoch=10, factor=10, maximum_lr=2048)
x, y = get_schedule(scheduler, epochs=50)
```

```python
plot(x, y)
```

## ExponentialLR

```python
scheduler = ExponentialLR(base_lr=256, from_epoch=10, factor=8, maximum_lr=4096)
x, y = get_schedule(scheduler, epochs=50)
```

```python
plot(x, y)
```

## HalvingLR

```python
scheduler = HalvingLR(base_lr=256, from_epoch=10, num_epochs=5, maximum_lr=2048)
x, y = get_schedule(scheduler, epochs=50)
```

```python
plot(x, y)
```

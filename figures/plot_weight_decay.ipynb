```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
names = ['fwd_decay_inv', 'lrn_decay_inv', 'train_acc', 'test_acc', 'conv_0', 'linear_0', 'conv_1', 'linear_1', 'conv_2', 'linear_2', 'conv_3', 'linear_3', 'conv_4', 'linear_4', 'conv_5', 'linear_5', 'conv_6', 'linear_6', 'linear_out']
df = pd.read_csv('weight_decay.csv', names=names, header=None)
layers_columns = df.columns[4:]
df
```

```python
df[layers_columns].var(axis=0)
```

```python
selected_block = 5
block_columns = [f'conv_{selected_block}', f'linear_{selected_block}']
variables = ['fwd_decay_inv', 'lrn_decay_inv']
```

```python
cols = variables + block_columns
df[cols]
```

```python
df['lrn_decay_inv'] == 0
```

```python
no_decay_placeholder = 25_000
df['fwd_decay_inv'] = df['fwd_decay_inv'].replace(0, no_decay_placeholder)
df['lrn_decay_inv'] = df['lrn_decay_inv'].replace(0, no_decay_placeholder)
df
```

```python
x_labels = df['fwd_decay_inv'].replace(no_decay_placeholder, 'No decay').unique()
y_labels = [no_decay_placeholder, 3000, 1000]
```

```python
params = {
   'font.size': 24,
   'axes.linewidth': 1.25,
   'axes.labelsize': 24,
   'legend.fontsize': 24,
   'xtick.labelsize': 24,
   'ytick.labelsize': 24,
}
plt.rcParams.update(params)
```

```python
title = 'Effects of weight decay on the convolutional layer'
x_label = '$\eta_{inv}$ forward'
y_label = 'Mean absolute value of the weights'
legend_labels = ['$\eta_{inv}^{lr}$ = ' + str(value) for value in y_labels]
legend_labels[0] = 'No decay'
```

```python
from palettable.colorbrewer.qualitative import Set2_3
colors = Set2_3.mpl_colors
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
# ax.set_title(title)
ax.set_xlabel(x_label)
ax.set_ylabel(y_label)

sns.lineplot(
    data=df, x='fwd_decay_inv',y=block_columns[0],
    hue='lrn_decay_inv', ax=ax, marker='o', errorbar=None, 
    linewidth=3, markersize=10, palette=colors
)

ax.invert_xaxis() 
plt.xticks(df['fwd_decay_inv'].unique(), labels=x_labels)
ax.tick_params(axis='x')
ax.tick_params(axis='y')  

legend = plt.legend(legend_labels)
frame = legend.get_frame()
frame.set_edgecolor('0.5')

plt.grid(linewidth=1, alpha=0.5, axis='both')
plt.tight_layout()
plt.savefig('weight_decay_effect.pdf')
```

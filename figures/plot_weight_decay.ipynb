```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weight_decay_effect.csv')
layers_columns = df.columns[4:]
df
```

```python
df[layers_columns].var(axis=0)
```

```python
selected_block = 2
block_columns = [f'conv_encoder_0{selected_block}', f'linear_pred_0{selected_block}']
variables = ['fwd_decay_inv', 'subnet_decay_inv']
```

```python
cols = variables + block_columns
df[cols]
```

```python
df['subnet_decay_inv'] == 0
```

```python
df['fwd_decay_inv'] = df['fwd_decay_inv'].replace(0, 10_000)
df['subnet_decay_inv'] = df['subnet_decay_inv'].replace(0, 3000)
df
```

```python
values = [250, 500, 1000, 3000]
df = df[df['subnet_decay_inv'].isin(values)]
```

```python
x_labels = df['fwd_decay_inv'].replace(10_000, 'No decay').unique()
y_labels = df['subnet_decay_inv'].unique()
```

```python
params = {
   'font.size': 24,
   'axes.linewidth': 1.25,
   'axes.labelsize': 24,
   'legend.fontsize': 24,
   'xtick.labelsize': 24,
   'ytick.labelsize': 24,
}
plt.rcParams.update(params)
```

```python
title = 'Effects of weight decay on the convolutional layer'
x_label = '$\eta_{inv}$ forward'
y_label = 'Mean absolute value of the weights'
legend_labels = ['$\eta_{inv}^{lr}$ = ' + str(value) for value in y_labels]
legend_labels[0] = 'No decay'
```

```python
from palettable.colorbrewer.qualitative import Set2_4
colors = Set2_4.mpl_colors
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
# ax.set_title(title)
ax.set_xlabel(x_label)
ax.set_ylabel(y_label)

sns.lineplot(
    data=df, x='fwd_decay_inv',y=block_columns[0],
    hue='subnet_decay_inv', ax=ax, marker='o', errorbar=None, 
    linewidth=3, markersize=10, palette=colors
)

ax.invert_xaxis() 
plt.xticks(df['fwd_decay_inv'].unique(), labels=x_labels)
ax.tick_params(axis='x')
ax.tick_params(axis='y')  

legend = plt.legend(legend_labels)
frame = legend.get_frame()
frame.set_edgecolor('0.5')
plt.grid(linewidth=1, alpha=0.5, axis='both')
plt.tight_layout()
plt.savefig('weight_decay_effect.pdf')
```
